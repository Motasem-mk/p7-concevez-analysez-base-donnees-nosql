# 1. Connect to MongoDB

```python
from pymongo import MongoClient
import polars as pl

# Connect using credentials from your Docker container
client = MongoClient("mongodb://127.0.0.1:27031")

# Select your database and collection
db = client["noscites"]
collection = db["Paris_listings"]
```

# 2. Load Data from MongoDB for Analysis

To perform the advanced statistical analysis required in this project, we extract relevant fields from the MongoDB `Paris_listings` collection using PyMongo.

We limit the selection to the fields needed for the following indicators:
- Type of housing (`room_type`)
- Availability (`availability_365`)
- Neighborhood (`neighbourhood_cleansed`)
- Host status (`host_is_superhost`)
- Reviews (`number_of_reviews`, `reviews_per_month`)

We then convert the query results into a Polars DataFrame to leverage its speed and efficiency when analyzing large volumes of data.


```python
#  Fetch exactly the fields we need
cursor = collection.find({},
    {
      "_id": 0,
      "reviews_per_month": 1,
      "number_of_reviews": 1,
      "room_type": 1,
      "neighbourhood_cleansed": 1,
      "host_is_superhost": 1
    }
)

# Convert to Polars DataFrame
df = pl.DataFrame(list(cursor))
```

```python
df
```

# 3. Cast & clean

**Short Summary of Data Casting & Null-Handling**  
- **Business-rule imputation:** treat any blank or null `reviews_per_month` as `0.0`, since “no data” here means “zero reviews.”  
- **Type enforcement:** cast `reviews_per_month` → `Float64` (after imputation) and `number_of_reviews` → `Int64` so all metrics compute correctly.  
- **Null preservation elsewhere:** keep nulls in other fields (e.g. `host_is_superhost`) when there’s no clear “default” value, surfacing true missing data for later schema or quality checks.

```python
df = df.with_columns([
    # Treat empty or null reviews_per_month as 0.0
    pl.col("reviews_per_month")
      .fill_null("")                  # turn actual nulls into ""
      .str.replace_all(r"^$", "0")    # empty-string → "0"
      .cast(pl.Float64)               # → 0.0 or actual float
      .alias("reviews_per_month"),

    # Cast number_of_reviews to integer
    pl.col("number_of_reviews")
      .cast(pl.Int64)
      .alias("number_of_reviews")
])
```

# 4. Average monthly reviews by room_type (nulls ignored automatically)


```python
avg_rate = (
    df
    .group_by("room_type")
    .agg([
        pl.col("reviews_per_month")
          .mean()
          .round(2)
          .alias("avg_reviews_per_month")
    ])
    .sort("avg_reviews_per_month", descending=True)
)

print("✔ Average booking rate by room type:\n", avg_rate)
```

# 5. Overall median number_of_reviews


```python
median_overall = (
   df
    .select(
        pl.col("number_of_reviews")
          .median()
          .alias("median_reviews")
    )
)
print("\n✔ Median reviews overall:\n", median_overall)
```

# 6. Median reviews by host category ("t" vs "f")

```python
median_by_super = (
    df
    .group_by("host_is_superhost")
    .agg([
        pl.col("number_of_reviews")
          .median()
          .alias("median_reviews")
    ])
)

print("\n✔ Median reviews by host category:\n", median_by_super)
```

# 7. Listing density by neighbourhood (count)

```python
density = (
    df
    .group_by("neighbourhood_cleansed")
    .agg([
        pl.count().alias("listing_count")
    ])
    .sort("listing_count", descending=True)
)

print("\n✔ Top 10 neighbourhoods by listing density:\n", density.head(10))
```

# 8. Top 5 neighbourhoods by avg monthly reviews


```python
top_nb = (
    df
    .group_by("neighbourhood_cleansed")
    .agg([
        pl.col("reviews_per_month")
          .mean()
          .round(2)
          .alias("avg_reviews_per_month")
    ])
    .sort("avg_reviews_per_month", descending=True)
    .head(5)
)
print("\n✔ Top 5 neighbourhoods by avg monthly reviews:\n", top_nb)
```


---

**Why we used Polars (vs. Pandas)**  
- **Speed & Parallelism:** Polars is written in Rust and automatically uses multiple threads, so large-group-by and aggregations finish far faster than single-threaded Pandas.  
- **Memory Efficiency:** Under the hood it uses Arrow-style columnar memory, so it holds big tables more compactly and without many temporary copies.  
- **Null-Safety & Strong Types:** Polars enforces clear types (Int64, Float64, Utf8) and has null-aware aggregators by default, making it easier to apply precise imputation rules.  
- **Lazy & Eager Modes:** You can build complex pipelines and let Polars optimize the execution graph, then collect at the end—something Pandas doesn’t offer.
